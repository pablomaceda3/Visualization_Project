```python
import pandas as pd
```

```python
df_filled_na = df.fillna(method='ffill',axis=0)
```

```python
df_filled_na
```

```python
df_dropped_columns = df_filled_na.drop(columns=['<5 Conﬁr..', '<5 Deaths', '<5 year H..', 'Deaths', 'GTS Target', 'Hospitalisation', 'MDG Target'])
```

```python
list(df_dropped_columns)
```

```python
df_renamed = df_dropped_columns.rename(columns={'Conﬁrmed Cases':'Confirmed Cases'})
```

```python
df_renamed
```

```python
df_renamed.drop(df_renamed.tail(1).index,inplace=True) 
```

```python
df_renamed['Confirmed Cases'] = df_renamed['Confirmed Cases'].apply(lambda x:str(x.replace(',','')))
```

```python
df_renamed['Confirmed Cases'] = pd.to_numeric(df_renamed['Confirmed Cases'])
```

```python
df_pivoted = df_renamed.pivot_table(values="Confirmed Cases",index="Year",columns="Country")
```

```python
df_pivoted
```

```python
df_pivoted.drop(df_pivoted.tail(2).index, inplace=True)
```

```python
df_pivoted
```

```python
df_pivoted.drop(columns='French Guiana', inplace=True)
```

```python
df_pivoted.to_csv("data_files/malaria_data.csv")
```
